##### function for plotting images


```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

def ImagePlot(img):
    ''' img is a np.asarray''' 
    imgplot = plt.imshow(img)
    return imgplot



### example:
main_path = '/home/michal_s/CGGProject/Spring23_CGG/data/raw/'
image_path = main_path+'1332.png'
img = np.asarray(Image.open(image_path))
ImagePlot(img)
    


```

##### function for resizing images


```python
import cv2
import matplotlib.pyplot as plt


def ImageResize(img,hight,width):
    ''' img is a csv2.imread/numpy.ndarray
    hight, width - use the desired parameters to resize into .'''
    resized = cv2.resize(img, dsize=(hight,width))
    return resized



### example:
main_path = '/home/michal_s/CGGProject/Spring23_CGG/data/raw/'
image_path = main_path+'1332.png'
img=cv2.imread(image_path)
resized_img= ImageResize(img,50,50)


print('Image Width, Height are: ',img.shape[1],img.shape[0])
print('resized Image Width, Height are: ',resized_img.shape[1],resized_img.shape[0])
imgplot = plt.imshow(resized_img)
imgplot
```
